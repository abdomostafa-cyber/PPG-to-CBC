```python
# List of notebooks to run

import os
notebooks = os.listdir('D:/simedy/Notebooks/CBC/')
notebooks = [notebook for notebook in notebooks if notebook.startswith('vital')][0]
notebooks
```

```python
import os
import shutil
import papermill as pm
from tqdm.notebook import tqdm

# Input task name
task_name = "Vital-modeling"

# Create a task-specific output folder
output_folder = f"output_{task_name}"
os.makedirs(output_folder, exist_ok=True)

# Run notebooks and save to the task-specific output folder
for notebook in tqdm(notebooks, desc="Running Notebooks"):
    output_notebook = os.path.join(output_folder, f"executed_{os.path.basename(notebook)}")
    for i in range(5):
        try:
            pm.execute_notebook(
                notebook,
                output_notebook
            )
            # print(f'Successfully executed {notebook}')
        except Exception as e:
            #Sometimes the notebooks won't keep going because some labs have only 1 class (Basophils and NRBCs)
            continue
    print(f"Notebook {notebook} has been executed 5 times!")

# Create legacy folder structure
legacy_folder = os.path.join('legacy_notebooks', task_name)
os.makedirs(legacy_folder, exist_ok=True)

# Move executed notebooks to the legacy folder
for notebook in tqdm(os.listdir(output_folder), desc="Moving Notebooks to Legacy Folder"):
    src = os.path.join(output_folder, notebook)
    dst = os.path.join(legacy_folder, notebook)
    shutil.move(src, dst)

# Remove the task-specific output folder if empty
if not os.listdir(output_folder):
    os.rmdir(output_folder)

print("All notebooks have been executed and moved to the legacy folder.")
```
